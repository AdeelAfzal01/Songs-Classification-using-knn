```python
import json
import pandas as pd
```

```python
with open('data/product_insight_data_v22.json', encoding='utf8') as f:
    data = json.load(f)
```

```python
with open('data/ProductDataCatalogueStaging.json', encoding='utf8') as f:
    d_data = json.load(f)
```

```python
len(d_data)
```

```python
# This functions adds 1 into frequency if current category sent by "data_iterator" function already in the dictionary
# else initialize the new category with 1

def data_counter(category):
    
    if isinstance(category, str):
        
        try: 
            
            data_frequency[category] = data_frequency[category] + 1
            
        except:
            
            data_frequency[category] = 1
```

```python
# A recursive function with base condition the current key is either a string \ float \ int. If the key is a dict or list
# then it'll keep continue calling itself.

def data_iterator(current_data, s = ""):
    
    if isinstance(current_data, dict):
        
        for key, value in current_data.items():
            
            if isinstance(value, str) or isinstance(value, float) or isinstance(value, int):
                
                data_counter(s+key)
                
            elif isinstance(value, list):
                
                if len(value) == 0:
                    
                    data_counter(s+key)
                    
                else:
                    
                    temp_s = s + key + '/'
                    
                    data_iterator(current_data[key][0], temp_s)
                    
                
            elif isinstance(value, dict):
                
                temp_s = s + key + '/'
                
                data_iterator(current_data[key], temp_s)
```

```python
## Product Insight Data
data_frequency = dict()

for d in data:
    
    data_iterator(d)
    
product_insight_data_frequency = data_frequency

```

```python
## Product Catalogue Staging Data

data_frequency = dict()

for d in d_data:
    
    data_iterator(d)
    
product_catalogue_staging_data_frequency = data_frequency
```

```python
# writing data into csv function

def write_data_to_csv(filename, data_dict):
    
    with open(filename, 'w', encoding='utf-8') as f:
        
        f.write("%s,%s\n"%("Category","Frequency"))
        
        for key, value in data_dict.items():
            
            f.write("%s,%d\n"%(key, value))
            
```

```python
# writing the product_insight data into a csv file.
write_data_to_csv("product_insight_data_frequency.csv", product_insight_data_frequency)

# writing the product_catalogue data into a csv file.
write_data_to_csv("product_catalogue_staging_data_frequency.csv", product_catalogue_staging_data_frequency)


```

```python
## removing the 'results/' fromt he product_insight data file.

temp_data = [(key.replace('results/', ''), value) if 'results/' in key else (key, value) for key, value in product_insight_data_frequency.items()]

updated_product_insight_data_frequency = dict()

for item in temp_data:
    
    updated_product_insight_data_frequency[item[0]] = item[1]
```

```python
## Summing the frequency count from product_insight data and product_catalogue data with common categories.

product_with_common_category_frequency = dict()

for key in updated_product_insight_data_frequency.keys():
    
    for k in product_catalogue_staging_data_frequency.keys():
        
        if (key == k):
            
            product_with_common_category_frequency[key] = updated_product_insight_data_frequency[key] + product_catalogue_staging_data_frequency[k]
        
        
```

```python
# writing the summed up frequences of common categories into a csv file.
write_data_to_csv("product_with_common_category_frequency.csv", product_with_common_category_frequency)
```
